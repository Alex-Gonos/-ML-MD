#LSTM Recurrent neural network simulating rugged oscillator

```python
import sys
sys.path.append('..')

import numpy as np
from matplotlib import pyplot as plt
from Code.dynamical_system import *
from Code.time_integrator import *
from Code.neuralnet import *

np.random.seed(2512517)
```

## Neural network integrator

Set system parameters, construct dynamical system and integrator.

The model system we are using here is the rugged oscillator.

The timestep size of the Neural network integrator takes values in $\Delta t_{\text{Verlet}},20\Delta t_{\text{Verlet}},40\Delta t_{\text{Verlet}},100\Delta t_{\text{Verlet}}$ where $\Delta t_{\text{Verlet}}$ is the step size of the Verlet integrator that is used for training.

```python
# mass of particle
mass = 1.2
# timestep for Verlet integrator
dt_verlet = 0.005
# timestep for neural network integrator
dt_1 = dt_verlet
dt_2 = 20*dt_verlet
dt_3 = 40*dt_verlet
dt_4 = 100*dt_verlet
# number of steps for neural network integrator
nsteps = 5

rugged = Rugged(mass)

verlet_rugged = VerletIntegrator(rugged,dt_verlet)

lstm_integrator_1 = LSTMFinalIntegrator(rugged,nsteps,dt_1)
lstm_integrator_2 = LSTMFinalIntegrator(rugged,nsteps,dt_2)
lstm_integrator_3 = LSTMFinalIntegrator(rugged,nsteps,dt_3)
lstm_integrator_4 = LSTMFinalIntegrator(rugged,nsteps,dt_4)

# visualise the neural network model
lstm_integrator_1.model.summary()
keras.utils.plot_model(lstm_integrator_1.model, show_shapes=True)
```

```python
BATCH_SIZE=64

data_generator_1 = DataGenerator(lstm_integrator_1,verlet_rugged)
train_batches_1 = data_generator_1.dataset.batch(BATCH_SIZE)

data_generator_2 = DataGenerator(lstm_integrator_2,verlet_rugged)
train_batches_2 = data_generator_2.dataset.batch(BATCH_SIZE)

data_generator_3 = DataGenerator(lstm_integrator_3,verlet_rugged)
train_batches_3 = data_generator_3.dataset.batch(BATCH_SIZE)

data_generator_4 = DataGenerator(lstm_integrator_4,verlet_rugged)
train_batches_4 = data_generator_4.dataset.batch(BATCH_SIZE)
```

## Train neural network based integrator

```python
EPOCHS=300
STEPS_PER_EPOCH=100
log_dir = './tb_logs/'
tensorboard_cb = tf.keras.callbacks.TensorBoard(log_dir=log_dir)

result_1 = lstm_integrator_1.model.fit(train_batches_1,epochs=EPOCHS,steps_per_epoch=STEPS_PER_EPOCH,
                                 callbacks=tensorboard_cb)
result_2 = lstm_integrator_2.model.fit(train_batches_2,epochs=EPOCHS,steps_per_epoch=STEPS_PER_EPOCH,
                                 callbacks=tensorboard_cb)
result_3 = lstm_integrator_3.model.fit(train_batches_3,epochs=EPOCHS,steps_per_epoch=STEPS_PER_EPOCH,
                                callbacks=tensorboard_cb)
result_4 = lstm_integrator_4.model.fit(train_batches_4,epochs=EPOCHS,steps_per_epoch=STEPS_PER_EPOCH,
                                 callbacks=tensorboard_cb)
```

## Visualise Loss function

```python
#print(result.history.keys())

plt.rcParams['figure.figsize'] = [10, 10]

# summarize history for loss
plt.plot(result_1.history['loss'])
plt.plot(result_2.history['loss'])
plt.plot(result_3.history['loss'])
plt.plot(result_4.history['loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.rc('font', size=12)          # controls default text sizes
plt.rc('axes', titlesize=20)     # fontsize of the axes title
plt.rc('axes', labelsize=20)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=20)    # fontsize of the tick labels
plt.rc('ytick', labelsize=20)    # fontsize of the tick labels
plt.rc('legend', fontsize=20)    # legend fontsize
plt.rc('figure', titlesize=20)  # fontsize of the figure title
plt.legend(['$\Delta_v$','$20\Delta_v$','$40\Delta_v$','$100\Delta_v$'], loc='upper left')
plt.yscale("log")
plt.show()
```

## Plot trajectories to $t=32$ generated by Verlet integrator and neural network based integrator

(Poorly performing baseline timestep separated for legibility)

```python
# Final time
T_final = 32

# Initial conditions
x = np.zeros(1)
v = np.zeros(1)
x[0] = -0.8
v[0] = -0.3
verlet_rugged.set_state(x,v)

# ==== Verlet integrator ====
t = 0.0
t_verlet_rugged = []
x_verlet_rugged = []
v_verlet_rugged = []
E_verlet_rugged = []



while t<T_final:
    t_verlet_rugged.append(t)
    E_verlet_rugged.append(verlet_rugged.energy())
    x_verlet_rugged.append(verlet_rugged.x[0])
    v_verlet_rugged.append(verlet_rugged.v[0])
    verlet_rugged.integrate(1)
    t += dt_verlet
        

# ==== Neural network integrator ====

# Initialise with Verlet integrator
x_initial_1 = np.zeros((lstm_integrator_1.nsteps,lstm_integrator_1.dynamical_system.dim))
v_initial_1 = np.zeros((lstm_integrator_1.nsteps,lstm_integrator_1.dynamical_system.dim))
x_initial_2 = np.zeros((lstm_integrator_2.nsteps,lstm_integrator_2.dynamical_system.dim))
v_initial_2 = np.zeros((lstm_integrator_2.nsteps,lstm_integrator_2.dynamical_system.dim))
x_initial_3 = np.zeros((lstm_integrator_3.nsteps,lstm_integrator_3.dynamical_system.dim))
v_initial_3 = np.zeros((lstm_integrator_3.nsteps,lstm_integrator_3.dynamical_system.dim))
x_initial_4 = np.zeros((lstm_integrator_4.nsteps,lstm_integrator_4.dynamical_system.dim))
v_initial_4 = np.zeros((lstm_integrator_4.nsteps,lstm_integrator_4.dynamical_system.dim))



verlet_rugged.set_state(x,v)
for k in range(lstm_integrator_1.nsteps):    
    x_initial_1[k,:] = verlet_rugged.x[:]
    v_initial_1[k,:] = verlet_rugged.v[:]
    verlet_rugged.integrate(int(dt_1/dt_verlet))
verlet_rugged.set_state(x,v)
for k in range(lstm_integrator_2.nsteps):    
    x_initial_2[k,:] = verlet_rugged.x[:]
    v_initial_2[k,:] = verlet_rugged.v[:]
    verlet_rugged.integrate(int(dt_2/dt_verlet))
verlet_rugged.set_state(x,v)
for k in range(lstm_integrator_3.nsteps):    
    x_initial_3[k,:] = verlet_rugged.x[:]
    v_initial_3[k,:] = verlet_rugged.v[:]
    verlet_rugged.integrate(int(dt_3/dt_verlet))
verlet_rugged.set_state(x,v)
for k in range(lstm_integrator_4.nsteps):    
    x_initial_4[k,:] = verlet_rugged.x[:]
    v_initial_4[k,:] = verlet_rugged.v[:]
    verlet_rugged.integrate(int(dt_4/dt_verlet))


lstm_integrator_1.set_state(x_initial_1,v_initial_1)
t_1 = (lstm_integrator_1.nsteps-1)*lstm_integrator_1.dt
lstm_integrator_2.set_state(x_initial_2,v_initial_2)
t_2 = (lstm_integrator_2.nsteps-1)*lstm_integrator_2.dt
lstm_integrator_3.set_state(x_initial_3,v_initial_3)
t_3 = (lstm_integrator_3.nsteps-1)*lstm_integrator_3.dt
lstm_integrator_4.set_state(x_initial_4,v_initial_4)
t_4 = (lstm_integrator_4.nsteps-1)*lstm_integrator_4.dt



# Timestepping loop
t_lstm_1 = []
x_lstm_1 = []
v_lstm_1 = []
E_lstm_1 = []
t_lstm_2 = []
x_lstm_2 = []
v_lstm_2 = []
E_lstm_2 = []
t_lstm_3 = []
x_lstm_3 = []
v_lstm_3 = []
E_lstm_3 = []
t_lstm_4 = []
x_lstm_4 = []
v_lstm_4 = []
E_lstm_4 = []


while t_1<T_final:
    t_lstm_1.append(t_1)
    x_lstm_1.append(lstm_integrator_1.x[0])
    v_lstm_1.append(lstm_integrator_1.v[0])
    E_lstm_1.append(lstm_integrator_1.energy())
    lstm_integrator_1.integrate(1)
    t_1 += dt_1
while t_2<T_final:
    t_lstm_2.append(t_2)
    x_lstm_2.append(lstm_integrator_2.x[0])
    v_lstm_2.append(lstm_integrator_2.v[0])
    E_lstm_2.append(lstm_integrator_2.energy())
    lstm_integrator_2.integrate(1)
    t_2 += dt_2
while t_3<T_final:
    t_lstm_3.append(t_3)
    x_lstm_3.append(lstm_integrator_3.x[0])
    v_lstm_3.append(lstm_integrator_3.v[0])
    E_lstm_3.append(lstm_integrator_3.energy())
    lstm_integrator_3.integrate(1)
    t_3 += dt_3
while t_4<T_final:
    t_lstm_4.append(t_4)
    x_lstm_4.append(lstm_integrator_4.x[0])
    v_lstm_4.append(lstm_integrator_4.v[0])
    E_lstm_4.append(lstm_integrator_4.energy())
    lstm_integrator_4.integrate(1)
    t_4 += dt_4
 



x_lstm_1 = np.array(x_lstm_1)
x_lstm_2 = np.array(x_lstm_2)
x_lstm_3 = np.array(x_lstm_3)
x_lstm_4 = np.array(x_lstm_4)
x_verlet_rugged = np.array(x_verlet_rugged)
v_lstm_1 = np.array(v_lstm_1)
v_lstm_2 = np.array(v_lstm_2)
v_lstm_3 = np.array(v_lstm_3)
v_lstm_4 = np.array(v_lstm_4)
v_verlet_rugged = np.array(v_verlet_rugged)


# Plot position error as a function of time

plt.rcParams['figure.figsize'] = [10, 10]

#plt.plot(t_lstm_1,x_lstm_1-x_verlet_rugged[int((nsteps-1)*dt_1/dt_verlet)::int(dt_1/dt_verlet)],label='$\Delta_v$',color='blue')
plt.plot(t_lstm_2,x_lstm_2-x_verlet_rugged[int((nsteps-1)*dt_2/dt_verlet):int(np.size(x_verlet_rugged)-dt_2/dt_verlet):int(dt_2/dt_verlet)],label='$20\Delta_v$',color='orange')
plt.plot(t_lstm_3,x_lstm_3-x_verlet_rugged[int((nsteps-1)*dt_3/dt_verlet)::int(dt_3/dt_verlet)],label='$40\Delta_v$',color='green')
plt.plot(t_lstm_4,x_lstm_4-x_verlet_rugged[int((nsteps-1)*dt_4/dt_verlet):int(np.size(x_verlet_rugged)-dt_4/dt_verlet):int(dt_4/dt_verlet)],label='$100\Delta_v$',color='red')
plt.legend(loc='upper left')
ax = plt.gca()
ax.set_xlabel('time $t$')
ax.set_ylabel('position Error $x_{NN}(t)-x_V(t)$')
ax.set_title('Position Error')
plt.rc('font', size=12)          # controls default text sizes
plt.rc('axes', titlesize=20)     # fontsize of the axes title
plt.rc('axes', labelsize=20)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=20)    # fontsize of the tick labels
plt.rc('ytick', labelsize=20)    # fontsize of the tick labels
plt.rc('legend', fontsize=20)    # legend fontsize
plt.rc('figure', titlesize=20)  # fontsize of the figure title
#plt.grid()
plt.show()

# Plot velocity error as a function of time

#plt.plot(t_lstm_1,v_lstm_1-v_verlet_rugged[int((nsteps-1)*dt_1/dt_verlet)::int(dt_1/dt_verlet)],label='$\Delta_v$',color='blue')
plt.plot(t_lstm_2,v_lstm_2-v_verlet_rugged[int((nsteps-1)*dt_2/dt_verlet):int(np.size(v_verlet_rugged)-dt_2/dt_verlet):int(dt_2/dt_verlet)],label='$20\Delta_v$',color='orange')
plt.plot(t_lstm_3,v_lstm_3-v_verlet_rugged[int((nsteps-1)*dt_3/dt_verlet)::int(dt_3/dt_verlet)],label='$40\Delta_v$',color='green')
plt.plot(t_lstm_4,v_lstm_4-v_verlet_rugged[int((nsteps-1)*dt_4/dt_verlet):int(np.size(v_verlet_rugged)-dt_4/dt_verlet):int(dt_4/dt_verlet)],label='$100\Delta_v$',color='red')
plt.legend(loc='upper left')
ax = plt.gca()
ax.set_xlabel('time $t$')
ax.set_ylabel('Velocity Error $v_{NN}(t)-v_V(t)$')
ax.set_title('Velocity Error')
plt.rc('font', size=12)          # controls default text sizes
plt.rc('axes', titlesize=20)     # fontsize of the axes title
plt.rc('axes', labelsize=20)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=20)    # fontsize of the tick labels
plt.rc('ytick', labelsize=20)    # fontsize of the tick labels
plt.rc('legend', fontsize=20)    # legend fontsize
plt.rc('figure', titlesize=20)  # fontsize of the figure title
#plt.grid()
plt.show()

# Plot L2 error as a function of time

#plt.plot(t_lstm_1,((v_lstm_1-v_verlet_rugged[int((nsteps-1)*dt_1/dt_verlet)::int(dt_1/dt_verlet)])**2+(x_lstm_1-x_verlet_rugged[int((nsteps-1)*dt_1/dt_verlet)::int(dt_1/dt_verlet)])**2)**0.5,label='$\Delta_v$',color='blue')
plt.plot(t_lstm_2,((v_lstm_2-v_verlet_rugged[int((nsteps-1)*dt_2/dt_verlet):int(np.size(v_verlet_rugged)-dt_2/dt_verlet):int(dt_2/dt_verlet)])**2+(x_lstm_2-x_verlet_rugged[int((nsteps-1)*dt_2/dt_verlet):int(np.size(x_verlet_rugged)-dt_2/dt_verlet):int(dt_2/dt_verlet)])**2)**0.5,label='$20\Delta_v$',color='orange')
plt.plot(t_lstm_3,((v_lstm_3-v_verlet_rugged[int((nsteps-1)*dt_3/dt_verlet)::int(dt_3/dt_verlet)])**2+(x_lstm_3-x_verlet_rugged[int((nsteps-1)*dt_3/dt_verlet)::int(dt_3/dt_verlet)])**2)**0.5,label='$40\Delta_v$',color='green')
plt.plot(t_lstm_4,((v_lstm_4-v_verlet_rugged[int((nsteps-1)*dt_4/dt_verlet):int(np.size(v_verlet_rugged)-dt_4/dt_verlet):int(dt_4/dt_verlet)])**2+(x_lstm_4-x_verlet_rugged[int((nsteps-1)*dt_4/dt_verlet):int(np.size(x_verlet_rugged)-dt_4/dt_verlet):int(dt_4/dt_verlet)])**2)**0.5,label='$100\Delta_v$',color='red')
plt.legend(loc='upper left')
ax = plt.gca()
ax.set_xlabel('time $t$')
ax.set_ylabel('L2 Error')
ax.set_title('L2 Error')
plt.rc('font', size=12)          # controls default text sizes
plt.rc('axes', titlesize=20)     # fontsize of the axes title
plt.rc('axes', labelsize=20)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=20)    # fontsize of the tick labels
plt.rc('ytick', labelsize=20)    # fontsize of the tick labels
plt.rc('legend', fontsize=20)    # legend fontsize
plt.rc('figure', titlesize=20)  # fontsize of the figure title
#plt.grid()
plt.show()


plt.clf()
# Plot energy as a function of time
# (subtract energy at time t=0 to show energy drift)

plt.plot(t_verlet_rugged,E_verlet_rugged-E_verlet_rugged[0],label='rugged',color='black')
#plt.plot(t_lstm_1,E_lstm_1-E_lstm_1[0],label='$\Delta_v$',color='blue')
plt.plot(t_lstm_2,E_lstm_2-E_lstm_2[0],label='$20\Delta_v$',color='orange')
plt.plot(t_lstm_3,E_lstm_3-E_lstm_3[0],label='$40\Delta_v$',color='green')
plt.plot(t_lstm_4,E_lstm_4-E_lstm_4[0],label='$100\Delta_v$',color='red')
plt.legend(loc='upper left')
ax = plt.gca()
ax.set_xlabel('time $t$')
ax.set_ylabel('energy shift $E(t)-E(t_0)$')
#ax.grid()
plt.rc('font', size=12)          # controls default text sizes
plt.rc('axes', titlesize=20)     # fontsize of the axes title
plt.rc('axes', labelsize=20)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=20)    # fontsize of the tick labels
plt.rc('ytick', labelsize=20)    # fontsize of the tick labels
plt.rc('legend', fontsize=20)    # legend fontsize
plt.rc('figure', titlesize=20)  # fontsize of the figure title

plt.show()

#subplots of position versus velocity for all 4

plt.rcParams['figure.figsize'] = [10, 10]

plt.clf()
plt.plot(x_verlet_rugged,v_verlet_rugged,label='Verlet',color='black')
#plt.plot(x_lstm_1,v_lstm_1,label='$\Delta_v$',color='blue')
plt.plot(x_lstm_2,v_lstm_2,label='$20\Delta_v$',color='orange')
plt.plot(x_lstm_3,v_lstm_3,label='$40\Delta_v$',color='green')
plt.plot(x_lstm_4,v_lstm_4,label='$100\Delta_v$',color='red')
plt.legend(loc='upper left')
ax = plt.gca()
ax.set_ylabel('Velocity $v(t)$')
ax.set_xlabel('Position $q(t)$')
plt.rc('font', size=12)          # controls default text sizes
plt.rc('axes', titlesize=20)     # fontsize of the axes title
plt.rc('axes', labelsize=20)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=20)    # fontsize of the tick labels
plt.rc('ytick', labelsize=20)    # fontsize of the tick labels
plt.rc('legend', fontsize=20)    # legend fontsize
plt.rc('figure', titlesize=20)  # fontsize of the figure title
plt.show()
```
